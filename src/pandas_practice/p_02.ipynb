```python
import os
import glob
import pandas as pd
```

```python
path_list = glob.glob("/workspaces/TIR102_Shi/res_gossiping/*.txt")
```

```python
data = []
for path in path_list:
    with open(path, "r") as f:
        article_string = f.read() # 整篇文章的字串
    reply_info_string = article_string.split("---split---")[1]
    reply_info_list = reply_info_string.split("\n")[1:-1]
    data.append(reply_info_list)
    # print(data)
    # print("====")

```

```python

columns = [
    "推",
    "噓",
    "分數",
    "作者",
    "標題",
    "時間",
]

df = pd.DataFrame(data=data, columns=columns)
df
```

```python
df.to_csv("ppt.csv", index=False)
```
